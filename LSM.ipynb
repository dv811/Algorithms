## Least Squares Method (LSM)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100 #number of values
sd=3 #standart diviation of values

#y=kx+l, where
k=0.5
l=5
```

```python
x=np.array(range(n)) #set x
f=np.array([k*i+l for i in range(n)]) #set f(x)
```

```python
y=f+np.random.normal(0,sd,n) #set y; loc =0, diviation=s, size=n
```

```python
plt.plot(f)
plt.scatter(x,y,s=5,c='red')
plt.grid(True)
```

```python
mx=x.sum()/n
my=y.sum()/n

a2=np.dot(x.T,x)/n
a11=np.dot(x.T,y)/n
```

```python
k_=(a11-mx*my)/(a2-mx**2)
l_=my-k_*mx
```

```python
f_=np.array([k_*i+l_ for i in range(n)])
```

```python
plt.plot(f_,c='black')
plt.plot(f)
plt.scatter(x,y,s=5,c='red')
plt.grid(True)
```

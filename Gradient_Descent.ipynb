# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x): 
    return x*x - 5*x + 5

def df(x):
    return 2*x - 5

n = 20   # number of iterations
x = 0
step = 0.1 

x_plt = np.arange(0, 5.0, 0.1)
f_plt = [f(x) for x in x_plt]

fig, ax = plt.subplots()    
ax.grid(True)

ax.plot(x_plt, f_plt)                  
point=ax.scatter(x, f(x), c='red')
```

```python
for i in range(n):
    x = x - step*df(x) # changing arg after each iteration
    point.set_offsets([x, f(x)]) 
print(x)
fig
```
